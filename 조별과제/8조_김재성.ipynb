```python
# 기후변화가 대구폭염(환경적변화)에 영향을 미칠까?
# 출처 : https://data.kma.go.kr/climate/heatWave/selectHeatWaveChart.do 기상청
# 대구폭염일수와 대구폭염기간 확인
# 대구폭염 일수와 대구폭염기간이 가장 길었던 해와 길지 않았던 해 찾기.
# 어떤 환경적 요인이 작용 했는지 설명하기

FILE=pd.read_csv('STCS_fokyum_count_day.csv', encoding='euc_kr')
FILE2=pd.read_csv('daegu_gin_fokyum.csv', encoding='euc_kr')

fokyum_count=pd.DataFrame(FILE)
fokyum_len=pd.DataFrame(FILE2)

```

```python
fokyum_count
```

```python
fokyum_len
```

```python
fokyum_count.drop(columns='순위', inplace=True)
fokyum_count
fockyum_count_list=[]
for a in fokyum_count['연합계']:
    fockyum_count_list.append(a)
```

```python

fokyum_len_list=[]
for b in fokyum_len['지속일수']:
    fokyum_len_list.append(b)
    
```

```python
year_list=[f'{year}년' for year in range(2014,2025)]
year_list
```

```python
print(year_list)
print(fockyum_count_list)
print(fokyum_len_list)
```

```python
plt.figure(figsize=(10,5))
plt.title('대구 폭염횟수 와 최대폭염지속일', size=16)

plt.plot(fockyum_count_list, label='대구폭염횟수')
plt.plot(year_list,fokyum_len_list, label='대구폭염지속일')
plt.legend()

plt.xlabel('년도')
plt.ylabel('DAY')
plt.show()
```
